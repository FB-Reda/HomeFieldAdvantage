```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(16,9))
plt.plot(dfn.HomePCT + np.random.normal(0,0.01,len(dfn)),dfn.AwayPCT + np.random.normal(0,0.01,len(dfn)),'o', alpha = 0.4)
plt.plot([0,1],[0,1])
plt.xlabel('Home PCT')
plt.ylabel('Away PCT')
plt.title('Home vs Away PCT in the NBA')
plt.savefig('Home vs Away PCT.png')
```

```python
dfn['Home_Factor'] = np.true_divide((dfn.HomePCT - dfn.AwayPCT),(dfn.HomePCT + dfn.AwayPCT))

import seaborn as sns
sns.kdeplot(dfn.Home_Factor, shade = True)
plt.xlim([-0.5,1.0])
plt.legend(['NBA'])
plt.plot([0,0],[0,3])
plt.xlabel('Home Factor')
```

```python
plt.plot(dfn.index.get_level_values(1) + np.random.normal(0,0.5,len(dfn)),dfn.Home_Factor,'o', alpha = 0.8)
x = dfn.index.get_level_values(1)
y = dfn.Home_Factor
fn = np.polyfit(x,y,1)
fit_fn = np.poly1d(fn) 
plt.plot(x,fit_fn(x))
plt.xlabel('Year')
plt.ylim([-1,1])
plt.xlim([1950,2010])
```
