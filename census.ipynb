```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('~/Documents/datamining/project/census-data0/ACS_16_5YR_S1903.csv', encoding="latin-1", skiprows=1)
df
```

```python
arr = {"Boston", "Philadelphia", "Chicago", "Cleveland-", "Detroit", "Indianapolis", "Milwaukee", "Atlanta", "Charlotte-", "Miami", "Orlando", "Washington", "Denver", "Minneapolis", "Oklahoma", "Portland-Vancouver", "Salt Lake", "Oakland", "Phoenix", "Sacramento", "Dallas", "Houston", "Memphis", "New Orleans", "San Antonio"}
df1 = df[["Geography", "Total; Estimate; Households", "Median income (dollars); Estimate; Households"]]
df1
```

```python
df2 = pd.DataFrame(index=range(3))
cities = pd.DataFrame(index=range(3))
cntr = 1

for city in arr:
    if cntr is 1:
        cities = df1[df1['Geography'].str.contains(city)]
        cntr = cntr + 1
    else:
        df2 = df1[df1['Geography'].str.contains(city)]
        cities = cities.append(df2)

cities = cities.sort_values('Geography', ascending = True)
cities
```

```python
nba = pd.read_csv('~/Documents/datamining/project/datasets/NBA.csv', encoding="latin-1")
nba
```

```python
for index, row in nba.iterrows():
    if row["Year"] is 2010:
        print("correct year")
```

```python
blank = True
blacklist = {'LosAngeles', 'Los Angeles', 'NewYork', 'New Jersey', 'NewJersey', 'New York', 'Toronto'}

for index, row in nba.iterrows():
    if row["Year"] == 2010:
        blacklisted = False
        for name in blacklist:
            if name in row['Team']:
                blacklisted = True
        if not blacklisted:
            if blank is True:
                combined = pd.DataFrame([[row['Team'],row['HomePCT'],row['TotalPCT'], row['HomePCT']-row['TotalPCT'], 0, 0]], columns = ['Team', 'HomePCT', 'TotalPCT', 'DifferentialPCT', 'MedianIncome', 'TotalHouseholds'])
                blank = False
            else:
                new = pd.DataFrame([[row['Team'],row['HomePCT'],row['TotalPCT'], row['HomePCT']-row['TotalPCT'], 0, 0]], columns = ['Team', 'HomePCT', 'TotalPCT', 'DifferentialPCT', 'MedianIncome', 'TotalHouseholds'])
                combined = combined.append(new)
        
combined
```

```python
metrodict = {'GoldenState' : 'Oakland', 'Minnesota' : 'Minneapolis', 'NewOrleans' : 'New Orleans', 'OklahomaCity' : 'Oklahoma', 'SanAntonio' : 'San Antonio', 'Utah' : 'Salt Lake'}
inclist = [0]
inclist.clear()
houselist = [0]
houselist.clear()

for index, row in combined.iterrows():
    teamplace = row['Team']
    firstword = teamplace.split('\xa0')[0]
    if firstword in metrodict:
        teamplace = metrodict[firstword]
    else:
        teamplace = firstword
    print(teamplace)
    for citindex, citrow in cities.iterrows():
        if teamplace in citrow['Geography']:
            print(citrow['Median income (dollars); Estimate; Households'])
            inclist.append(citrow['Median income (dollars); Estimate; Households'])
            houselist.append(citrow['Total; Estimate; Households'])
    print()
            
len(inclist)
len(combined['MedianIncome'])

combined['MedianIncome'] = inclist
combined['TotalHouseholds'] = houselist
```

```python
combined
```

```python
combined.plot(kind='scatter', y='DifferentialPCT', x='MedianIncome', s=30)
```

```python
regression0 = np.polyfit(combined['MedianIncome'], combined['DifferentialPCT'], 1)
print(regression0)
```

```python
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='DifferentialPCT ~ MedianIncome', data=combined).fit()

# print the coefficients
lm.params
```

```python
-0.030271 + 0.000002*100000
```

```python
52654.547161 + 80801.603400*.05
```

```python
X_0 = pd.DataFrame({'MedianIncome' : [combined.MedianIncome.min(), combined.MedianIncome.max()]})
X_0
```

```python
preds0 = lm.predict(X_0)
preds0
```

```python
combined.plot(kind='scatter', x='MedianIncome', y='DifferentialPCT')
plt.plot(X_0, preds0, c='red', linewidth=2)
```

```python
lm.conf_int()
```

```python
lm.rsquared
```
